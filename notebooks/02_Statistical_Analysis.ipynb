# Statistical Analysis

This notebook explores statistical relationships between AI adoption, automation, revenue growth, productivity, and employee satisfaction.

Objectives:

1. Identify correlations between key variables.
2. Test whether differences between AI maturity groups are statistically significant.
3. Build regression models to quantify relationships.

Dataset Size:

1,000,000 rows
25 columns

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
df = pd.read_csv('../data/cleaned_ai_workforce.csv')
```

# Correlation Analysis: 
Correlation analysis helps identify whether AI adoption and automation are associated with changes in productivity, revenue growth, and employee satisfaction.

While correlation does not imply causation, it can highlight relationships worth investigating further through hypothesis testing and regression analysis.

The purpose of this test is to see correlations between: 

- AI Adoption 
- Automation 
- Revenue Growth 
- Productivity 
- Employee Satisfaction 

We can use this to see relationships between these variables. 

```python
corr = df[[
    'ai_adoption_score',
    'automation_level',
    'revenue_growth',
    'productivity_index',
    'employee_satisfaction'
]].corr()

corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(corr,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Correlation Matrix')
plt.show()
```

### Correlation Findings

Key observations:

- AI Adoption and Automation show a very strong positive correlation (0.91).
- AI Adoption has a moderate positive relationship with Productivity (0.55).
- AI Adoption has a relatively weak positive relationship with Revenue Growth (0.22).
- Productivity and Employee Satisfaction show a moderate positive relationship (0.47).

These findings suggest that AI adoption may influence business outcomes indirectly through productivity improvements rather than directly driving revenue growth.

# Hypothesis Testing

Business Question:

Do companies with Leading AI adoption achieve significantly higher revenue growth than companies with Minimal AI adoption?

Statistical Test:

Independent Samples T-Test

Null Hypothesis (H₀):
There is no difference in average revenue growth between Leading and Minimal AI adoption companies.

Alternative Hypothesis (H₁):
There is a significant difference in average revenue growth between Leading and Minimal AI adoption companies.

```python
leading = df[df['ai_adoption_tier']=='Leading']['revenue_growth']
minimal = df[df['ai_adoption_tier']=='Minimal']['revenue_growth']

#hypothesis testing
t_stat, p_value = ttest_ind(
    leading,
    minimal,
    equal_var=False
)

results = {
    't_statistic': t_stat,
    'p_value': p_value,
    'leading_mean': leading.mean(),
    'minimal_mean': minimal.mean()
}

results

```

```python
df.groupby('ai_adoption_tier')['revenue_growth'].agg(
    ['count','mean','median','std']
)
```

## Results

T Statistic: 109.60

P Value: < 0.001

Decision:

Reject the null hypothesis.

Key Findings:

- Leading AI adopters averaged 11.04% revenue growth.
- Minimal AI adopters averaged 2.93% revenue growth.
- Revenue growth increased consistently across all AI adoption tiers.

Additional Observation:

Mean and median revenue growth values were very similar across all groups. This suggests that the observed relationship is not driven by a small number of extreme outliers.

Conclusion:

Higher levels of AI adoption are associated with significantly higher revenue growth.

# Productivity Analysis

Business Question:

Do companies with Leading AI adoption achieve significantly higher productivity than companies with Minimal AI adoption?

Statistical Test:

Independent Samples T-Test

Null Hypothesis (H₀):

There is no difference in average productivity between Leading and Minimal AI adoption companies.

Alternative Hypothesis (H₁):

There is a significant difference in average productivity between Leading and Minimal AI adoption companies.

```python
leading_prod = df[df['ai_adoption_tier'] == 'Leading']['productivity_index']

minimal_prod = df[df['ai_adoption_tier'] == 'Minimal']['productivity_index']

t_stat, p_value = ttest_ind(
    leading_prod,
    minimal_prod,
    equal_var=False
)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```

```python
productivity_summary = (
    df.groupby('ai_adoption_tier')['productivity_index']
      .agg(['count', 'mean', 'median', 'std'])
)

productivity_summary
```

## Results

T Statistic: 353.89

P Value: < 0.001

Decision:

Reject the null hypothesis.

Key Findings:

- Leading AI adopters achieved an average productivity score of 83.54 compared to 63.76 for Minimal adopters.
- Productivity increased consistently across all AI adoption tiers.
- Mean and median values were very similar across all groups, suggesting the relationship is not driven by extreme outliers.

Conclusion:

Companies with higher levels of AI adoption demonstrate significantly higher productivity scores.

# Linear Regression Analysis

Business Question:

How much does AI adoption score predict revenue growth?

Purpose:

The previous analysis showed that revenue growth increases across AI adoption tiers. Linear regression helps estimate the relationship between the continuous AI adoption score and revenue growth.

Model:

Revenue Growth = β₀ + β₁(AI Adoption Score)

Interpretation:

- The coefficient for AI Adoption Score shows the expected change in revenue growth for a one-point increase in AI adoption score.
- R-squared shows how much variation in revenue growth is explained by AI adoption score alone.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X = df[['ai_adoption_score']]
y = df['revenue_growth']
```

```python
model = LinearRegression()
model.fit(X, y)

intercept = model.intercept_
coefficient = model.coef_[0]

print("Intercept:", intercept)
print("Coefficient:", coefficient)
```

```python
y_pred = model.predict(X)

r2 = r2_score(y, y_pred)

print("R-squared:", r2)
```

## Results

Intercept: 1.45

Coefficient: 0.12

R-Squared: 0.046

Key Findings:

- AI adoption score has a positive relationship with revenue growth.
- For every 1-point increase in AI adoption score, revenue growth increases by approximately 0.12 percentage points.
- AI adoption explains approximately 4.6% of variation in revenue growth.

Conclusion:

AI adoption appears to be a meaningful predictor of revenue growth. However, most variation in revenue growth is explained by other business and economic factors not included in this model.

# Final Conclusions

Key findings from this analysis:

- AI adoption is strongly associated with automation levels.
- Companies with higher AI adoption demonstrate significantly higher revenue growth.
- Companies with higher AI adoption demonstrate significantly higher productivity.
- Linear regression indicates a positive relationship between AI adoption and revenue growth, although AI adoption explains only a portion of revenue growth variation.
